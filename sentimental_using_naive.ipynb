```python
import pandas as pd
docs = pd.read_csv('Covid_tweets_US.csv',low_memory=False)
#classifier in column 1, sms in column 2.
docs.head()
```

```python
docs['label'] = docs.Polarity.map({'Negative':2, 'Positive':1,'Neutral':0})
```

```python
docs.head()
```

```python
X = docs.text
y = docs.label
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection  import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.20, random_state =42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')
```

```python
 vect.fit(X_train.values.astype('U'))
vectorizer = vect.fit(X.values.astype('U'))
```

```python
vect.vocabulary_
```

```python
print(vect.get_feature_names())
print(len(vect.get_feature_names()))
```

```python
X_train_transformed = vect.transform(X_train.values.astype('U'))
X_test_transformed =vect.transform(X_test.values.astype('U'))
```

```python
print(type(X_train_transformed))
print(X_train_transformed)
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()

# fit
naive_model = mnb.fit(X_train_transformed,y_train)

# predict class
y_pred_class = mnb.predict(X_test_transformed)

# predict probabilities
y_pred_proba =mnb.predict_proba(X_test_transformed)

print(y_pred_proba)
```

```python
from sklearn import metrics

confusion = metrics.confusion_matrix(y_test, y_pred_class)
print(confusion)
#[row, column]
TP = confusion[0, 0]
FN = confusion[0, 1]
FP = confusion[1, 0]
TN = confusion[1, 1]
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_class)
```

```python
x = metrics.confusion_matrix(y_test, y_pred_class)
x
```

```python
sensitivity = TP / float(FN + TP)
print("sensitivity",sensitivity)
```

```python
specificity = TN / float(TN + FP)

print("specificity",specificity)
```

```python
precision = TP / float(TP + FP)

print("precision",precision)
print(metrics.precision_score(y_test, y_pred_class,average='weighted'))
```

```python
print("precision",precision)
print("PRECISION SCORE :",metrics.precision_score(y_test, y_pred_class,average='weighted'))
print("RECALL SCORE :", metrics.recall_score(y_test, y_pred_class,average='weighted'))
print("F1 SCORE :",metrics.f1_score(y_test, y_pred_class,average='weighted'))
```
